## Exploratory Data Analysis

> Use Matplotlib to display your data analysis

### Which production companies released the most movies in the last 10 years? Display the top 5 production companies.

```python
prod = df[df['release_year'] >=2005].groupby('production_companies').count()[['original_title']]

prod.sort_values( 'original_title', ascending=False).head(5)

```

### What 5 movie genres grossed the highest all-time?

```python
gen = df.groupby('genres').sum()[['revenue']].sort_values('revenue', ascending=False).head()
gen
```

### Who are the top 5 grossing directors?

```python
richdirector = df.groupby('director').sum()[['revenue']].sort_values('revenue', ascending = False).head()
richdirector
```

### Compare the revenue of the highest grossing movies of all time.

```python
movie = df.groupby(['original_title', 'genres', 'release_year', 'director']).sum()[['revenue']]
movie.sort_values('revenue', ascending=False).head(20)
```
